```python
from pyspark.sql import SparkSession

# Create Spark Session
spark = SparkSession.builder \
    .appName("Big Data Analysis Task") \
    .getOrCreate()

print("Spark Session Created Successfully")
```

```python
# Load dataset from online source
data = spark.read.csv(
    "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv",
    header=True,
    inferSchema=True
)

# Show first 5 rows
data.show(5)
```

```python
# Print schema (structure)
data.printSchema()

# Count total records
print("Total Records:", data.count())
```

```python
data.groupBy("day") \
    .avg("total_bill") \
    .show()
```

```python
data.groupBy("sex") \
    .avg("tip") \
    .show()
```

```python
data.groupBy("smoker") \
    .avg("total_bill") \
    .show()
```

# INSIGHTS DERIVED FROM BIG DATA ANALYSIS:

1. Average total bill varies across different days of the week.
2. Non-smokers tend to have slightly higher average bills.
3. Tipping behavior shows variation between genders.
4. PySpark efficiently handled aggregation and grouping operations,
   demonstrating scalability for big data processing.
